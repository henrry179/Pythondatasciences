## MATLAB风格接口
- Matplotlib最初作为MATLAB用户的Python替代品，许多语法都和MATLAB类似。
- MATLAB风格的工具位于pyplot(plt)接口中。

```python
%matplotlib inline
# 创建图形
plt.figure()

# 创建两个子图中的第一个，设置坐标轴
plt.subplot(2, 1, 1)     # (行、列、子图编号)
plt.plot(x, np.sin(x))

# 创建两个子图中的第二个，设置坐标轴
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
plt.show()
```

### 面向对象接口

```python
# 先创建图形网格
# ax是一个包含两个Axes对象的数组
fig, ax = plt.subplots(2)

# 在每个对象上调用plot()方法
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
plt.show()
```

### 调整图形：线条的颜色与风格

```python
plt.plot(x, np.sin(x - 0), color = 'blue')             # 标准颜色名称
plt.plot(x, np.sin(x - 1), color = 'g')                # 缩写颜色代码(rgbcmyk)
plt.plot(x, np.sin(x - 2), color = '0.75')             # 范围在0~1的灰度值
plt.plot(x, np.sin(x - 3), color = '#FFDD44')          # 十六进制(RRGGBB, 00-FF)
plt.plot(x, np.sin(x - 4), color = (1.0, 0.2, 0.3))    # RGB元组，范围在0~1
plt.plot(x, np.sin(x - 5), color = 'chartreuse');      # HTML颜色名称
plt.show()
```

```python
plt.plot(x, x + 0, linestyle = 'solid')
plt.plot(x, x + 1, linestyle = 'dashed')
plt.plot(x, x + 2, linestyle = 'dashdot')
plt.plot(x, x + 3, linestyle = 'dotted')

# 还可以用下面的简写形式
plt.plot(x, x + 4, linestyle = '-')      # 实线
plt.plot(x, x + 5, linestyle = '--')     # 虚线
plt.plot(x, x + 6, linestyle = '-.')     # 点划线
plt.plot(x, x + 7, linestyle = ':');     # 实点线
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\dashdot.pdf")
```

```python
如果想用一种更简洁的方式，则可以将linestyle和color编码组合起来，作为plt.plot()函数的一个非关键字参数使用
```

```python
# 不同风格的线条
plt.plot(x, x + 0, 'g')        # 绿色实线
plt.plot(x, x + 1, '--c')     # 青色虚线
plt.plot(x, x + 2, '-.k')      # 黑色点划线
plt.plot(x, x + 3, ':r')       # 红色实点线
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\plot1.pdf")
```

```python
### 调整图形：坐标轴上下限
虽然Matplotlib会自动为你的图形选择最合适的坐标轴上下限，但是有时自定义坐标轴上下限可能会更好。调整坐标轴上下限最基础的方法是plt.xlim()和plt.ylim().
```

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\lim.pdf")
```

```python
还有一个方法是plt.axis()，通过传入[xmin, xmax, ymin, ymax]对应的值，plt.axis()方法可以让你用一行代码设置x和y的限值。
```

```python
# 坐标轴刻度值逆序
plt.plot(x, np.sin(x))

plt.xlim(10 ,0)
plt.ylim(1.2, -1.2);
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\lim2.pdf")
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\lim3.pdf")
```

```python
plt.axis()能做的可不止如此，它还可以按照图形的内容自动收紧坐标轴，不留空白区域。
```

```python
# plt.axis()设置坐标轴上限
plt.plot(x, np.sin(x))
plt.axis('tight')
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\tight1.pdf")
```

```python
# “相等”布局示例，分辨率为 1:1
plt.plot(x, np.sin(x))
plt.axis('equal')
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\equal.pdf")
```

```python
### 设置图形标签
- 图形标题
- 坐标轴标题
- 简易图例
```

```python
# 图形标题与坐标轴标题
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\sine_curve.pdf")
```

```python
#### 图例
- Matplotlib内置了一个简单快速的方法，可以用来创建图例，那就是plt.legend().
- 虽然有不少用来设置图例的办法，但我还是觉得在plt.plot()函数中用label参数为每条线设置一个标签最简单。
```

```python
# 图例
plt.plot(x, np.sin(x), 
         '-g', 
         label = 'sin(x)')

plt.plot(x, 
         np.cos(x), 
         ':b', 
         label = 'cos(x)')

plt.axis('equal')
plt.legend();
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\equal.pdf")
```

```python
在用面向对象接口画图时，不需要单独调用这些函数，通常采用ax.set()方法一次性设置所有的属性是更简便的方法。
```

```python
# 用ax.set()方法一次性设置所有的属性
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim = (0, 10), 
       ylim = (-2, 2), 
       xlabel = 'x', 
       ylabel = 'sin(x)', 
       title = 'A Simple Plot');

plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\simple_plot.pdf")
```

```python
## 简易散点图
另一种常用的图形是简易散点图(scatter plot),与线形图类似。这种图不再由线段连接，而是由独立的点、圆圈或其他形状构成。开始的时候同样需要在Notebook中导入函数：
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
### 用plt.plot画散点图
```

```python
# 散点图
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color = 'black');
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\black.pdf")
```

```python
#### 不同的图形标记
- 函数的第三个参数是一个字符，表示图形符号的类型。
- 与之前用'-'和'--'设置线条属性类似，对应的图形标记也有缩写形式。
- 所有的缩写形式都可以在plt.plot文档中查到。
- 绝大部分图形标记都非常直观。
```

```python
# 不同的图形标记
rng = np.random.RandomState(0)
for marker in ['o', 
               '.', 
               ',', 
               'x', 
               '+', 
               'v', 
               '^', 
               '<', 
               '>', 
               's', 
               'd']:
    
    plt.plot(rng.rand(5), 
             rng.rand(5), 
             marker, 
             label = "marker = '{0}''".format(marker))
    
    plt.legend(numpoints = 1)
    plt.xlim(0, 1.8);
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\numpoints.pdf")
```

```python
# 组合线条与散点
plt.plot(x, y, '-ok');       # 直线(-)、 画圈(o)、黑色(k)
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\plt3.pdf")
```

```python
# 自定义线条和散点属性
plt.plot(x, y, '-p', 
         color = 'gray', 
         markersize = 15, 
         linewidth = 4, 
         markerfacecolor = 'white', 
         markeredgecolor = 'gray', 
         markeredgewidth = 2)

plt.ylim(-1.2, 1.2);
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\markersize.pdf")
```

```python
### 用plt.scatter画散点图
```

```python
# 简易散点图
plt.scatter(x, y, marker = 'o');
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\scatter.pdf")
```

```python
# 改变散点图中散点的大小、颜色和透明度
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, 
            y, 
            c = colors, 
            s = sizes, 
            alpha = 0.3, 
            cmap = 'viridis')

plt.colorbar();   # 显示颜色条
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\scatter1.pdf")
```

```python
- 可以用Scikit-Learn程序库里面的鸢尾花(iris)数据来演示。它里面有三种鸢尾花，每个样本是一种花。
- 其花瓣(petal)与花萼(sepal)的长度与宽度都经过了仔细测量。
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], 
            features[1], 
            alpha = 0.2, 
            s = 100*features[3], 
            c = iris.target, 
            cmap = 'viridis')

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
plt.savefig(r"D:\Python-Excel\Data-analysis\数据分析实战技能\images\scatter.pdf")
```
